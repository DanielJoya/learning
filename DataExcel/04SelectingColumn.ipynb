## Adding a new columns with an array

The dataframe has 1000 rows, so we need to create and array with 1000 elements 

```python
# Import numpy
import numpy as np
```

```python
# Create and array of 1000 elements
language_score = np.arange(0,1000)
```

```python
# Length of the array
len(language_score)
```

```python
# Adding a new column to dataframe with an array
df_exams['langauge score'] = language_score
df_exams
```

```python
# Cretae random integer numbers between 1 and 100
int_language_score = np.random.randint(1,100, size=1000)
```

```python
# Min value inclusive and high value exclisive
min(int_language_score)
max(int_language_score)
```

```python
# Adding a new column to dataframe with an array
df_exams['langauge score'] = int_language_score
df_exams
```
